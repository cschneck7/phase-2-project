<strong>This notebook is for making figures with the intention of use in the presentation for aesthetics. They do not reflect the data.</strong> 

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fast')
%matplotlib inline
```

```python
# Sets coordinates for lines and titles

x = [[0, 1], [0,.75], [0,1]]
y = [[0,.7], [0,1], [0,.5]]
titles=['Feature A', 'Feature B', 'Feature C']
colors=['b', 'g', 'r']
features=['a', 'b', 'c']
formula=['y = ' + str(round((y[i][1]-y[i][0])/(x[i][1]-x[i][0]), 1)) + features[i] for i in range(3)]
```

```python
# Creates vertical plot of example figures

fig, axes = plt.subplots(figsize=(4,12), nrows=3, ncols=1)

for i in range(3):
    axes[i].plot(x[i], y[i], colors[i], linewidth=3)
    axes[i].set_xlim(-.1, 1.1)
    axes[i].set_ylim(-.1, 1.1)
    axes[i].set_title(titles[i])
    axes[i].xaxis.set_tick_params(labelbottom=False)
    axes[i].yaxis.set_tick_params(labelleft=False)
    fig.tight_layout(pad=5.0)
  
fig.savefig('ex_fig_vert.png', dpi=100)
```

```python
# Creqates horizontal plots of example figures

fig, axes = plt.subplots(figsize=(12,3), nrows=1, ncols=3)

for i in range(3):
    axes[i].plot(x[i], y[i], colors[i], linewidth=3)
    axes[i].set_xlim(-.1, 1.1)
    axes[i].set_ylim(-.1, 1.1)
    axes[i].set_title(titles[i])
    axes[i].xaxis.set_tick_params(labelbottom=False)
    axes[i].yaxis.set_tick_params(labelleft=False)
    axes[i].text(.1, .9, formula[i], fontsize=15)
#     fig.tight_layout(pad=5)
    
fig.savefig('ex_fig_hor_wtext.png', dpi=100)
```
